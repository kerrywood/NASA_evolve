```python
from poliastro.twobody import Orbit
from poliastro.bodies import Earth

import numpy as np
import astropy.time
import astropy.units as u

import pandas as pd

from sgp4.earth_gravity import wgs72
```

```python
NOW = astropy.time.Time.now()

def generateOrbit( I ):
    print(I,end='\r')
    a    = np.random.uniform( 200,1500 ) + wgs72.radiusearthkm 
    ecc  = np.random.uniform(1e-4,0.05)
    inc  = np.random.uniform(0,90)
    raan = np.random.uniform(0,360)
    argp = np.random.uniform(0,360)
    nu   = np.random.uniform(0,360)
    return Orbit.from_classical( Earth,
                                 a * u.km,
                                 ecc * u.one,
                                 inc * u.deg,
                                 raan * u.deg,
                                 argp * u.deg,
                                 nu * u.deg,
                                 epoch = NOW )

orbits = [generateOrbit( X ) for X in range(100000)]
```

```python
pos = np.vstack( [X.r.to_value(u.km) for X in orbits] )
vel = np.vstack( [X.v.to_value(u.km/u.s) for X in orbits] )
scc = np.arange( len(pos), dtype=np.int64) + 1
jd  = np.ones( len(pos) ) * NOW.jd 
sv = np.hstack( (pos,vel) )
todf = np.hstack( (scc[:,np.newaxis], jd[:,np.newaxis], sv ) )
```

```python
outdf = pd.DataFrame( todf, columns = ['scc','jd','gcrsx','gcrsy','gcrsz','gcrsdx','gcrsdy','gcrsdz'])
```

```python
outdf.to_csv('./100k_random_orbits.csv.gz')
```
